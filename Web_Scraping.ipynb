# Web Scraping - Codezella

```python
import requests # calls web pages
from bs4 import BeautifulSoup # library for web scraping
import csv
```

```python
date = input("Please enter a date (M/D/YYYY): ")
```

```python
url = f'https://www.yallakora.com/match-center/%d9%85%d8%b1%d9%83%d8%b2-%d8%a7%d9%84%d9%85%d8%a8%d8%a7%d8%b1%d9%8a%d8%a7%d8%aa?date={date}'
page = requests.get(url)
```

```python
source = page.content # Returns page content in bytes code
print(source)
```

In order to make the content understandable we parse the byte code to and get html or xml content. Which is similar to when we click inspect in google chrome. <br />
https://www.crummy.com/software/BeautifulSoup/bs4/doc/

```python
soup = BeautifulSoup(source, 'lxml')
```

```python
print(soup) # more readable compared to byte code
```

If we go and inspect the html in the web page, we will find html tags (such as div tag). we need to locate the parent tag of which we need to access using beautiful soup. <br />
Championship names are under the tag : \<div class="2745 matchCard matchesList"> (which is the parent we are looking for the targeted list) <br /><br />
BeautifulSoup find_all() will be used to filter the html. So instead of having full page html we will only preserve the parts we need.  

```python
matches_details = []

championship = soup.find_all("div", {'class': 'matchCard'}) #(tag, {class: classname})
```

```python
print(championship)
```

To get the championship name we need to look into \<h2> under div class "title" under div matchCard. <br />
to access tag's childern we need to use .contents from beautiful soup, which works as a list where each child of a parent class is represented by index number. <br /> a good practice is to check the contents index one by one.


```python
# title is the first class under div matchCard hence its index is 0
# we used find as we are looking for a single result
#.text looks for the text inside h2
#.strip removes spaces before and after the text
# contents[0] returned wrong values hence we try [1] and check if it contains the class title. 

championship_title = championship[0].contents[1].find('h2').text.strip()
print(championship_title)
```

```python
# championship index may vary depending on how many matchCard and title classes that we have in the webpage. 
championship_title = championship[1].contents[1].find('h2').text.strip()
print(championship_title)
```

```python
# list item 'li' includes information about matches and can be found in index 3 in contents
all_matches = championship[0].contents[3].find_all('li')

#each list item in all_matches belong to one match
no_of_matches = len(all_matches)
print(no_of_matches)
# 6 matches in the first championship 
```

```python
# Specify the championship
for j in range(len(championship)): 
    
    championship_title = championship[j].contents[1].find('h2').text.strip()
    all_matches = championship[j].contents[3].find_all('li')
    no_of_matches = len(all_matches)
    print(championship_title, no_of_matches)
    
    for i in range(no_of_matches):
        # get teams names
        team_A = all_matches[i].find('div', {'class': 'teamA'}).text.strip()
        team_B = all_matches[i].find('div', {'class': 'teamB'}).text.strip()

        #get score
        match_result = all_matches[i].find('div', {'class': 'MResult'}).find_all('span', {'class': 'score'})

        #we have 3 span under MResult span 0 for team a span 1 for team b and span 2 for the time
        score = f"{match_result[0].text.strip()} - {match_result[1].text.strip()} "

        #get match time
        time = all_matches[i].find('div', {'class': 'MResult'}).find('span', {'class': 'time'}).text.strip()
                
        #add match info to matches_details
        matches_details.append({"": championship_title,
                               " ": team_A,
                               " ": team_B,
                               "": score,
                               "": time})
        
```

```python
print(matches_details)
```

We have created list of dictionaries

```python
# get class names for columns
# we only need the headers once hence we will take it from the first dictionary only
keys = matches_details[0].keys()
print(keys)
```

```python
with open('P:\Programs\Jupyter files\web_scraping.csv', 'w', encoding='utf-8-sig') as output_file: #error without the encoding
    dict_writer = csv.DictWriter(output_file, keys) #this is a class
    dict_writer.writeheader() #this is a method that fills column names
    dict_writer.writerows(matches_details)
    print('File successfully created')
```
