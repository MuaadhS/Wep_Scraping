```python
import requests
from bs4 import BeautifulSoup
```

```python
url = f'https://features.salla.sa/ideas/?category=6995111627065623089&status=6994764301156281394'
page = requests.get(url)
```

```python
source = page.content # Returns page content in bytes code
print(source)
```

```python
soup = BeautifulSoup(source, 'lxml')
#print(soup)
```

```python
ideas = soup.find_all("h2", {"class" : "idea-name"})
#print(ideas)
description = soup.find_all("div", {"class" : "description"})
#print(description)
votes = soup.find_all("span", {"class" : "vote-count"})
links = soup.find_all("a", {"class" : "idea-link"})
```

```python
idea_name = []
idea_description = []
idea_votes = []
idea_link = []
```

```python
for i in range(len(ideas)):
        idea_name.append(ideas[i].text.strip())
        idea_link.append("https://features.salla.sa/" + links[i].get('href'))
        idea_description.append(description[i].text.strip())
        idea_votes.append(votes[i].text.strip())
```

```python
category = []
li_elements = soup.find_all("li")
for li in li_elements:
    li_id = li.get('id')
    if li_id:
        category.append(li_id)
```

```python
cleaned_category = [item.replace('category-', '') for item in category]

print(cleaned_category)
```

# Loop code

```python
idea_cat = []
idea_name = []
idea_description = []
idea_votes = []
idea_link = []

def loop(category, name, pg):
  p = 1
  while p<=pg:
    print(f"getting data from page{p} of catergory {name}")

    url = f'https://features.salla.sa/ideas/?category={category}&page={p}'
    page = requests.get(url)
    source = page.content
    soup = BeautifulSoup(source, 'lxml')

    ideas = soup.find_all("h2", {"class" : "idea-name"})
    description = soup.find_all("div", {"class" : "description"})
    votes = soup.find_all("span", {"class" : "vote-count"})
    links = soup.find_all("a", {"class" : "idea-link"})


    for i in range(len(ideas)):
      idea_cat.append(name)
      idea_name.append(ideas[i].text.strip())
      idea_link.append("https://features.salla.sa/" + links[i].get('href'))
      idea_description.append(description[i].text.strip())
      idea_votes.append(votes[i].text.strip())


    p +=1
```
